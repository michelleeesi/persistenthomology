```python
import glob
import scipy.io
import numpy as np
import pandas as pd
```

```python
dataDir = '/content/'
mats = []
d = {}
vertices_list_all = []
count =0;
for filename in glob.glob('**/*.mat*'): 
    count +=1;
# filename = 'res/decorated_tree.aca.Case100.mat'
    print('filename', filename)
    print('in the loop')
    subject_id = filename.split('.',2)[2]
    data = scipy.io.loadmat(filename)
    # mats.append(scipy.io.loadmat(filename))
    vals = data['T'][0,0] #<-- set the array you want to access. 
    keys = data['T'][0,0].dtype.descr

    # Assemble the keys and values into variables with the same name as that used in MATLAB
    for i in range(len(keys)):
        key = keys[i][0]
        val = np.squeeze(vals[key][0][0])  # squeeze is used to covert matlat (1,n) arrays into numpy (1,) arrays. 
        exec(key + '=val')

    # perhaps the coordinates of vertices
    list_coordinates_vertices = []
    for i in range(0,vals[1].shape[1]):
        list_coordinates_vertices.append(vals[1][0][i][0][0].tolist())
    #     print(vals[1][0][i][0][0])    
    vertices_list_all.append(list_coordinates_vertices)
    d[subject_id] = vertices_list_all
print(count);
```

```python
dict_order = list(d.keys())
```

```python
df = pd.DataFrame(dict_order, columns=['subject'])
```

```python
df['vertices_coordinates'] = vertices_list_all
```
